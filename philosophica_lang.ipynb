Symbol:

In der Philosophie beziehungsweise der Ästhetik ein Erkennungszeichen, einfach in der Form, reich und tief im Sinn. Auf Denk- und Grabmälern nicht unüblich. 

Beispiel: Der „Lindenbaum“ im Lied Am Brunnen vor dem Tore von Franz Schubert/Wilhelm Müller. Nach Dietrich Ritschl sind Symbole „Produkte bewusster, reifer Erkenntnisleistung durch Repräsentanzen in Form von Worten, Handlungen oder Gesten. Symbole vermitteln, was anders nicht artikuliert werden kann.“
    
    Index = Name des Symbols
    Weight = Gewichtung / Häufigkeit (zB -100, zB 42)
    

```python
import matplotlib.pyplot as  plt


class Symbol:
    s = '';
    weight = 0;
    def set(self,symbol,weight):
        self.s = symbol;
        self.weight = weight;
    def plot(self):        
        plt.plot( (0,self.weight), (-self.weight,0) )
        plt.annotate(self.s, (0,self.weight))
    def sum(self,s1):        
        ds = Symbol()
        d = s1.s + '.' + self.s;
        dw = s1.weight + self.weight;
        ds.set(d,dw);
        plt.plot( (0,s1.weight), (-self.weight,0) )
        return ds;
```

```python
associative = Symbol()
associative.set("Associative", 100)

dissociative = Symbol()
dissociative.set("Dissociative", -100)

alien = Symbol()
alien.set("Alien", 9999)

nachbar = Symbol()
nachbar.set("der große Andere", 140)

```

```python
associative.plot();
dissociative.plot();
nachbar.plot()
```
